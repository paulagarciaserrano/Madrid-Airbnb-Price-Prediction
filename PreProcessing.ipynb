# Dummies

```python
df = pd.get_dummies(df, prefix='room_type', columns=['room_type'], drop_first=True)
df = pd.get_dummies(df, prefix='bed_type', columns=['bed_type'], drop_first=True)
df = pd.get_dummies(df, prefix='cancellation_policy', columns=['cancellation_policy'], drop_first=True)
df = pd.get_dummies(df, prefix='area_name', columns=['area_name'], drop_first=True)
df = pd.get_dummies(df, prefix='host_response_time', columns=['host_response_time'], drop_first=True)
```

```python
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('-','_')
df.columns = df.columns.map(lambda x: x.lower())
```

# Saving as CSV

```python
df.to_csv('./CSVs/Cleaned_df.csv', index=False)
```
